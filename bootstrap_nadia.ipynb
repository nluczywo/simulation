```python
import garpar as gp
import itertools as it
import pandas as pd
import numpy as np
import joblib as jb
import tqdm.notebook
import scipy 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
merval21 = gp.datasets.load_MERVAL(
    first="01-02-2021", last = "12-31-2021")
merval21 = merval21.as_returns()
merval21
```

```python
# definimos los sectores 
come = ["AUSO", "BOLT", "DYCA","GCLA", "IRCP", "IRSA", "OEST", "PATA", "POLL","TECO2","CVH"] 
moa = ["CELU", "GRIM", "HAVA", "LEDE", "MOLA", "MOLI", "MORI", "SAMI", "SEMI"] #"INTR"
moi = ["AGRO", "ALUA",  "FERR", "LONG", "RIGO", "HARG", "FIPL"] # "ROSE", "LOMA", "CAPU","MIRG","TXAR "
agro = ["CADO", "GARO", "INVJ","CRES"] #"ESME"
ene = ["CEPU", "CGPA2", "DGCU2", "GBAN", "METR", "TGNO4", "TGSU2", "YPFD", "PAMP","TRAN","COME","EDN","GBAN" ]  #"CARC",
fin = ["BHIP", "BPAT",  "BYMA", "GGAL", "SUPV", "BBAR"] #"BMA", "VALO", "BRIO","BRIO6",
```

```python
alphacome = []
betacome = []
loccome = []
scalecome = []
pvalue = []
for i in range(2):
    sample_merval21= merval21.sample(frac=0.8, replace=True, random_state=i)
    y21scomes = sample_merval21[come].mean(axis = 1, skipna = True) 
    alpha_come, beta_come, loc_come, scale_come = scipy.stats.levy_stable.fit(y21scomes)
    ktestcom = scipy.stats.ks_1samp(x=y21scomes, cdf=scipy.stats.levy_stable.cdf, args=(alpha_come, beta_come, loc_come, scale_come ))
    alphacome.append(alpha_come)
    betacome.append(beta_come)
    loccome.append(loc_come)
    scalecome.append(scale_come) 
    pvalue.append(ktestcom.pvalue)
alpha_come = pd.Series(alphacome).mean(),
beta_come = pd.Series(betacome).mean(),
loc_come = pd.Series(loccome).mean(),
scale_come = pd.Series(scalecome).mean(),
average_pvalue = pd.Series(pvalue).mean(),
alpha_come, beta_come, loc_come, scale_come, average_pvalue
```

```python
alpha_interval = scipy.stats.t.interval(0.95, df=len(alphacome)-1, loc=alpha_come, scale=scipy.stats.sem(alphacome))
beta_interval = scipy.stats.t.interval(0.95, df=len(betaacome)-1, loc=beta_come, scale=scipy.stats.sem(betacome))
mu_interval = scipy.stats.t.interval(0.95, df=len(loccome)-1, loc=loc_come, scale=scipy.stats.sem(loccome))
sigma_interval =scipy.stats.t.interval(0.95, df=len(scalecome)-1, loc=scale_come,scale=scipy.stats.sem(scalecome))

alpha_interval, beta_interval, mu_interval
```

```python
%%time

def get_stats(df, rstate,sector):
    sample= df.sample(frac=0.8, replace=True, random_state=rstate)
    
    scomes = sample[sector].mean(axis = 1, skipna = True) 
    alpha_come, beta_come, loc_come, scale_come = scipy.stats.levy_stable.fit(scomes)
    
    ktestcom = scipy.stats.ks_1samp(
        x=y21scomes, 
        cdf=scipy.stats.levy_stable.cdf, 
        args=(alpha_come, beta_come, loc_come, scale_come)
    )
    
    return alpha_come, beta_come, loc_come, scale_come, ktestcom.pvalue


alphacome = []
betacome = []
loccome = []
scalecome = []
pvalue = []

for iteration in range(2):
    
    alpha_come, beta_come, loc_come, scale_come, ktestcom = get_stats(df=merval21, rstate=iteration, sector=come)
    
    alphacome.append(alpha_come)
    betacome.append(beta_come)
    loccome.append(loc_come)
    scalecome.append(scale_come) 
    pvalue.append(ktestcom)
    
alpha_come = np.mean(alphacome)
beta_come = np.mean(betacome)
loc_come = np.mean(loccome)
scale_ccome = np.mean(scalecome)
average_pvalue = np.mean(pvalue)
alpha_come, beta_come, loc_come, scale_ccome, average_pvalue
```
